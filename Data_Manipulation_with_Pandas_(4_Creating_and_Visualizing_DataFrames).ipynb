```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'City': ['New York', 'London', 'Paris', 'Tokyo'],
        'IsStudent': [True, False, True, False],
        'JoinDate': pd.to_datetime(['2023-01-15', '2023-02-20', '2023-03-10', '2023-04-05'])}

df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt


df['Age'].hist(bins=5)
plt.show()
```

```python
# Bar Plot -- can reveal relationship between categorical variable and numerical variable

is_students = df.groupby('IsStudent')['Age'].mean()
print(is_students)



is_students.plot(kind='bar',title='Average Age by Student Status')
plt.show()
```

```python
# Line plot
# is great for visualize changes in numeric variables over time


is_students.plot(x='Student',y='age',kind='line',rot=45)
plt.show()
```

# Dictionary

```python
# Creating DataFrame

# 1. From a list of dictionaries
list_of_dict = [{'Name': 'Alice', 'Age': 25, 'City': 'New York'},
                {'Name': 'Bob', 'Age': 30, 'City': 'London'},
                {'Name': 'Charlie', 'Age': 35, 'City': 'Paris'}]

df_from_list = pd.DataFrame(list_of_dict)
print(df_from_list)
```

```python
# 2. Dictionary of lists - by column

dict_of_lists = {'Name': ['Alice', 'Bob', 'Charlie'],
                 'Age': [25, 30, 35],
                 'City': ['New York', 'London', 'Paris']}

df_from_dict = pd.DataFrame(dict_of_lists)
print(df_from_dict)
```
