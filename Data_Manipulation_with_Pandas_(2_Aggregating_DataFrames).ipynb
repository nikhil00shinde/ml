```python
import pandas as pd
df = pd.DataFrame({
    'name':['john','mary','peter','jeff','bill','lisa','jose'],
    'age':[23,78,22,19,45,33,20],
    'gender':['M','F','M','M','M','F','M'],
    'salary'  : [50000,52000,56000,62000,37000,45000,52000]
})
```

```python
df['age'].mean() # .median(), .mode(), .min(), .max(), .var(), .std(), .sum()
```

```python
## .agg() --> The .agg() method allows you to apply your own custom functions to a DataFrame, as well as apply functions to more than one column of a DataFrame at once, making your aggregations super-efficient.
def pct30(column):
  return column.quantile(0.3)

def pct40(column):
  return column.quantile(0.4)

df['age_pct30'] = df['age'].agg(pct30)

## More than one column
df[['age', 'age_pct30']].agg(pct30)
df[['age', 'age_pct30']].agg([pct30,pct40,"median"])
```

```python
df['age'].cumsum() ## Like prefix sum
## .cumsum(), .cummin(), .cumprod()
```

## Counting

```python
print(df)
print(df.drop_duplicates(subset=['gender','name'])) # Get unique name and with gender
```

```python
# count the value of each gender and also we can sort
print(df['gender'].value_counts(sort=True))

# Get proportion
print(df[['gender']].value_counts(normalize=True))
```

```python
## Summaries by group
# df[df['age'] >= 30]
# df[df['age'] >= 50]['age'].mean()



# ## EASY WAY
df.groupby('gender')['age'].mean()

## Multiple grouped summaries
df.groupby('gender')['age'].agg([min,max,sum])


## Group by multiple multiple variables
df.groupby(['gender','age'])['age'].mean()
```

## Pivot Tables

```python
# same as groupby, default mean

# dogs.groupby('color')['weight_kg'].mean()

# dogs.pivot_table(values='weight_kg',index='color')


df.pivot_table(values='age',index='gender')

df.pivot_table(values='age',index='gender',aggfunc=['mean','median'])




# Pivot on two variables
df.pivot_table(values='age',index='gender',columns='salary',fill_value=0,margins=True)
```
