```python
## Pandas built using numpy + matplotlib

## Store data for analysis (Rectangular/Tabular --> known as DataFrame in python

```

```python
import pandas as pd

df = pd.DataFrame({
    'name': ['a','b','c','d'],
    'course': ['ml','dl','daa','db'],
    'grade': ['A','A','C','A']
})

print(df.head()) ## return first rows of the data
```

```python
print(df.info()) # display name of column
```

```python
print(df.shape) #show row, column length
```

```python
print(df.describe()) # summary statistics mean mode
```

```python
print(df.values) # 2d numpy array
print(df.columns) # return column names
print(df.index) # return index
```

## Sorting and Subsetting

```python
df.sort_values('grade',ascending=False)
df.sort_values(['grade','name'],ascending=[True,False])
```

```python
df[['name','grade']] ## Subsetting multiple columns
below_A = df['grade'] > 'A' ## Subsetting rows
df[below_A] # filtering
```

```python
df[(df['grade'] == 'A') & (df['name'].isin(['a','b']))] # categorise the value
```
